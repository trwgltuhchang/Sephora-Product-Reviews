# Filtering Products

```python
import pandas as pd

filtered_df5 = df5[
    (df5['review_count'] >= 10) &
    (df5['recommended_count'] > 0) &
    (df5['recommended_ratio'] >= 0.5) &
    (df5['price_usd'] <= 500)  # Optional price cap
]

removed_df5 = df5[~df5.index.isin(filtered_df5.index)]

# Output some summary
print(f"Original product count: {len(df5)}")
print(f"Filtered product count: {len(filtered_df5)}")
print(f"Removed products: {len(removed_df5)}")

filtered_df5.head(10)
```

```python
duplicate_check = filtered_df5.duplicated(subset=["brand_name", "product_name", "price_usd"], keep=False)

print("Number of duplicate rows:", duplicate_check.sum())

filtered_df5[duplicate_check].sort_values(by=["brand_name", "product_name", "price_usd"])
```

# The table to use ML

```python
#check relevant review texts
relevant_reviews = df_reviews.merge(
    filtered_df5[["brand_name", "product_name", "price_usd"]],
    on=["brand_name", "product_name", "price_usd"],
    how="inner"
)
```

```python
relevant_reviews = relevant_reviews[[
    "brand_name", "product_name", "price_usd", "review_title", "review_text", "helpfulness", "rating"
]]
```

```python
missing_titles = relevant_reviews["review_title"].isna().sum()
print(f"Missing review titles: {missing_titles}")
```

```python
#check random
columns_to_keep = ["brand_name", "product_name", "price_usd", "review_title", "review_text"]

reviews = (
    relevant_reviews[columns_to_keep]
    .groupby(["brand_name", "product_name", "price_usd"], group_keys=False)
    .apply(lambda x: x.sample(min(len(x), 50), random_state=10))
    .reset_index(drop=True)
)
```

```python
final_reviews = pd.merge(
    reviews,
    df_reviews,
    on=["brand_name", "product_name", "price_usd", "review_title", "review_text"],
    how="left"
)
print(final_reviews.columns)
final_reviews
```

# The use of ML for the **prediction** if a product is recommended

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from textblob import TextBlob
```

```python
# Sentiment score
final_reviews["sentiment"] = final_reviews["review_text"].fillna("").apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
# Prepare ML data
ml_df = final_reviews[["is_recommended", "rating", "helpfulness", "sentiment", "price_usd", "skin_type", "hair_color", "eye_color"]]
ml_df = pd.get_dummies(ml_df, drop_first=True)
```

```python
# Train/test split
X = ml_df.drop("is_recommended", axis=1)
y = ml_df["is_recommended"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Training
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)

# Plot
plt.figure(figsize=(6, 4))
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()
```
